```python
capacity1 = 4
capacity2 = 3
goal = 2

visited = set()

def dfs(x, y, steps):
    if x == goal or y == goal:
        print("Solution found:\n")
        for s in steps:
            print(s)
        print(f"\nFinal state: ({x}, {y})")
        return True

    visited.add((x, y))

    next_states = [
        ((capacity1, y), "Fill Jug 1"),
        ((x, capacity2), "Fill Jug 2"),
        ((0, y), "Empty Jug 1"),
        ((x, 0), "Empty Jug 2"),
        (pour_1_to_2(x, y), "Pour Jug1 → Jug2"),
        (pour_2_to_1(x, y), "Pour Jug2 → Jug1")
    ]

    for (new_x, new_y), action in next_states:
        if (new_x, new_y) not in visited:
            if dfs(new_x, new_y, steps + [action + f" -> ({new_x}, {new_y})"]):
                return True

    return False

def pour_1_to_2(x, y):
    space = capacity2 - y
    transfer = min(x, space)
    return (x - transfer, y + transfer)

def pour_2_to_1(x, y):
    space = capacity1 - x
    transfer = min(y, space)
    return (x + transfer, y - transfer)

dfs(0, 0, [])
```
