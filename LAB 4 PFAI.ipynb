```python
def solve_n_queens(n):
    board = [-1] * n
    solutions = []

    def is_safe(row, col):
        for r in range(row):
            if board[r] == col or abs(board[r] - col) == abs(r - row):
                return False
        return True

    def place_queen(row):
        if row == n:
            solutions.append(board.copy())
            return
        for col in range(n):
            if is_safe(row, col):
                board[row] = col
                place_queen(row + 1)
                board[row] = -1

    place_queen(0)
    return solutions


def print_solutions(solutions, n):
    print("Total solutions:", len(solutions))
    for sol in solutions:
        for c in sol:
            print(" ".join("Q" if i == c else "." for i in range(n)))
        print()


n = int(input("Enter N for N-Queens: "))
solutions = solve_n_queens(n)
print_solutions(solutions, n)
```
